```python
#LotArea
quantile_25 = np.quantile(df_train['LotArea'], 0.25)
quantile_75 = np.quantile(df_train['LotArea'],0.75)
IQR = quantile_75 - quantile_25

minimum = quantile_25 - 1.5 * IQR
maximum = quantile_75 + 1.5 * IQR
train_IQR = df_train[(minimum <= df_train['LotArea']) & (df_train['LotArea'] <= maximum)]
```

```python
print(train_IQR.shape)
print(df_train.shape[0]-train_IQR.shape[0])
```

```python
#
quantile_25 = np.quantile(train_IQR['MSSubClass'], 0.25)
quantile_75 = np.quantile(train_IQR['MSSubClass'],0.75)
IQR = quantile_75 - quantile_25

minimum = quantile_25 - 1.5 * IQR
maximum = quantile_75 + 1.5 * IQR
train_IQR = train_IQR[(minimum <= train_IQR['MSSubClass']) & (train_IQR['MSSubClass'] <= maximum)]
```

## GridSearch사용
1. fit하면 grid search를 사용하여 자동으로 복수개의 내부 모형 생성
2. 생성한 모형을 실행하여 최적 파라미터 검색.

```python
X_train = new_df.drop(columns=['SalePrice'])
Y_train = new_df['SalePrice']
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(criterion='mse')
```

```python
model.fit(X_train, Y_train)
```

```python
model.feature_importances_
```

```python
#YearRemodAdd제거
X_train = X_train.drop(columns=['YearRemodAdd'])
test_1 = test_df.drop(columns=['YearRemodAdd'])
```

```python
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor(criterion = 'mse', random_state=2022)

params = {'n_estimators': [200, 300, 500],
          'max_features': [4, 5, 6],
          'min_samples_leaf': [1, 3, 5]}

greedy_CV = GridSearchCV(model, param_grid=params, cv = 3, n_jobs = -1)
greedy_CV.fit(X_train, Y_train)
```

```python
pred = greedy_CV.predict(test_1)
```
